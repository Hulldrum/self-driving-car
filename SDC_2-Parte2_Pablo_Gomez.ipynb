## 2. Conversión a escala de grises en openCV 


```python
img = cv.imread('figuras/carretera.jpeg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
plt.show()
```

## 3. Aplicar el filtro Gaussiano

```python
img = cv.imread('figuras/carretera.jpeg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gray,(3,3),0)
plt.imshow(img_gauss,cmap='gray')
plt.show()
```

## 4. Aplicar el Thresholding


```python
img = cv.imread('figuras/carretera.jpeg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gray,(3,3),0)
thr, img_thr= cv.threshold(img_gauss ,108 ,255,cv.THRESH_BINARY)
plt.imshow( img_thr,cmap='gray')
plt.show()
```

## 5. Disminuir el tamaño de la altura

```python
img = cv.imread('figuras/carretera.jpeg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gray,(3,3),0)
thr, img_thr= cv.threshold(img_gauss ,108 ,255,cv.THRESH_BINARY)
alto=img.shape[0]
ancho=img.shape[1]
ratio=0.2
img_r = cv.resize(img_thr,(int(ancho*0.09),int(alto*0.09)), interpolation=cv.INTER_NEAREST)
plt.imshow( img_r,cmap='gray')
plt.show()
```

## 6. Todas las operaciones

```python
img = cv.imread('figuras/carretera.jpeg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gray,(3,3),0)
thr, img_thr= cv.threshold(img_gauss ,108 ,255,cv.THRESH_BINARY)
alto=img.shape[0]
ancho=img.shape[1]
ratio=0.2
img_r = cv.resize(img_thr,(int(ancho*0.09),int(alto*0.09)), interpolation=cv.INTER_NEAREST)

plt.figure(figsize=(10,6))
plt.subplot(2,3,1)
plt.imshow(img)
plt.title("Imagen original")
plt.subplot(2,3,2)
plt.imshow(img_gray,cmap='gray')
plt.title("Imagen en escala de grises")
plt.subplot(2,3,3)
plt.imshow(img_gauss, cmap='gray')
plt.title("Imagen con filtro Gaussiano")
plt.subplot(2,3,4)
plt.imshow(img_thr, cmap='gray')
plt.title("Imagen con Thresholding")
plt.subplot(2,3,5)
plt.imshow(img_r, cmap='gray')
plt.title("Imagen con el ratio = " + str(ratio))
plt.show()
```
