# 0. Librerías básicas

```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
import pandas as pd
```

# Actividad 1

```python
#Obtenga una imagen (libro, revista, etc..) por medio de su celular 
img = cv.imread("figuras/maths.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
#Ampliar y mejorar la visibilidad del plano cartesiano
fig, ax = plt.subplots(figsize=(14,14))
ax.imshow(img)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth='0.9', color='red')
ax.grid(which='minor', linestyle=':', linewidth='0.5', color='white')
```

```python
#Puntos de las esquinas
img = cv.imread("figuras/maths.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.figure(figsize=(14,14))
cv.circle(img, (540,750), 25, (0,0,150), -1) ;
cv.circle(img, (1900,200), 25, (0,0,150), -1) ;
cv.circle(img, (1100,2870), 25, (0,0,150), -1) ;
cv.circle(img, (3030,2015), 25, (0,0,150), -1) ;
plt.imshow(img)
plt.show()
```

```python
#Obteniendo la matriz de transformación y cambiando de perspectiva
pts1 = np.float32([[540,750], [1900,200], [1100,2870], [3030,2015]])
pts2 = np.float32([[0, 10], [1200, 0], [0, 2050], [1200, 2000]])
matrix = cv.getPerspectiveTransform(pts1, pts2)
img_warp = cv.warpPerspective(img, matrix, (1200,2000))
print("forma de matrix: ", matrix.shape)
print("forma de img_warp: ", img_warp.shape)
```

```python
#Imagen final 
plt.figure(figsize=(10,7))
plt.imshow(img_warp)
plt.grid()
plt.show()
```
